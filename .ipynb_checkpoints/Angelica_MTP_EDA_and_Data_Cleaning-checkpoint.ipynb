# Sleep Efficiency Dataset

## Day 1: EDA and Data Cleaning



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import math
from sklearn.preprocessing import PowerTransformer
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

import statsmodels.api as sm
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_absolute_error as mae
pd.options.display.max_rows = 50
import warnings
warnings.filterwarnings('ignore')
```

```python
# Loading the data base
sleep_df = pd.read_csv('Sleep_Efficiency.csv')
```

```python
# Checking general info: data types, number of columns.
sleep_df.info()
```

```python
# Eliminating duplicates
sleep_df = sleep_df.drop_duplicates()
sleep_df.info()
# No row was droppped, therefore there are no duplicates
```

```python
# Changing title of columns
sleep_df.rename(columns={'Wakeup time':'wakeup_time','Sleep Duration':'sleep_duration',
                          'Sleep efficiency':'sleep_efficiency','REM sleep percentage':'rem_sleep_percentage', 
                          'Deep sleep percentage':'deep_sleep_percentage','Light sleep percentage': 'light_sleep_percentage',
                          'Caffeine consumption':'caffeine_consumption','Alcohol consumption':'alcohol_consumption',
                         'Smoking status':'smoking_status','Exercise frequency':'exercise_frequency'}, inplace=True )
sleep_df
```

```python
# Making all columns to be in lower case
sleep_df = sleep_df.rename(columns= lambda x: x.lower())
sleep_df
```

```python
# Finding null entries in all columns
sleep_df.isnull().sum()
```

```python
# Creating a function that finds the mean of a column, leaving out the null values
def mean_nonzero(column_name):
    mean_list_nonzero = []
    for i in sleep_df[column_name]:
        if i is not None and not pd.isna(i):
            mean_list_nonzero.append(i)
    if len(mean_list_nonzero) > 0:
        mean = sum(mean_list_nonzero) / len(mean_list_nonzero)
        rounded_mean = round(mean, 2) 
        return rounded_mean
    else:
        return None
```

```python
# Appliying the function to find the mean in the columns where are null values
mean_awakenings = mean_nonzero('awakenings')
mean_caffeine_consumption = mean_nonzero('caffeine_consumption')
mean_alcohol_consumption = mean_nonzero('alcohol_consumption')
mean_exercise_frequency = mean_nonzero('exercise_frequency')
print ('mean_awakenings: ', mean_awakenings, \
      'mean_caffeine_consumption: ', mean_caffeine_consumption, \
      'mean_alcohol_consumption: ', mean_alcohol_consumption, \
      'mean_exercise_frequency: ', mean_exercise_frequency)
```

```python
# Using a lambda function to replace the null values for the mean
sleep_df['awakenings'] = \
sleep_df['awakenings'].apply(lambda x: mean_awakenings if x is None or pd.isna(x) else x)
```

```python
sleep_df['caffeine_consumption'] = \
sleep_df['caffeine_consumption'].apply(lambda x: mean_alcohol_consumption if x is None or pd.isna(x) else x)
```

```python
sleep_df['alcohol_consumption'] = \
sleep_df['alcohol_consumption'].apply(lambda x: mean_alcohol_consumption if x is None or pd.isna(x) else x)
```

```python
sleep_df['exercise_frequency'] = \
sleep_df['exercise_frequency'].apply(lambda x: mean_exercise_frequency if x is None or pd.isna(x) else x)
```

```python
# checking that all null values where replaced and there are not any more null values
sleep_df.isnull().sum()
```

```python
# To do:
 # bins para wake up and bed time
 # change integers to floats
```

```python
# Split the 'bedtime' column into date and time columns
sleep_df[['bedtime_date', 'bedtime_time']] = sleep_df['bedtime'].str.split(' ', 1, expand=True)
sleep_df[['wakeup_date', 'wakeuptime_time']] = sleep_df['wakeup_time'].str.split(' ', 1, expand=True)
sleep_df
```

```python
sleep_df = sleep_df.drop(columns=["bedtime","wakeup_time"])
sleep_df
```

```python
# Extracting the month of the date columns
# Convert the column to datetime if it's not already in datetime format
sleep_df['bedtime_date'] = pd.to_datetime(sleep_df['bedtime_date'])
# Extract only the month from the datetime column
sleep_df['month'] = sleep_df['bedtime_date'].dt.month
sleep_df = sleep_df.drop(columns=["bedtime_date","wakeup_date"])
sleep_df
```

```python
# checking the Dtype of the month columns
sleep_df.info()
```

```python
# Converting the hour into hours - integers for bedtime_time

# Convert the time column to datetime data type
sleep_df['bedtime_time'] = pd.to_datetime(sleep_df['bedtime_time'], format='%H:%M:%S')

# Extract the hour, minute, and second components
sleep_df['hour'] = sleep_df['bedtime_time'].dt.hour
sleep_df['minute'] = sleep_df['bedtime_time'].dt.minute
sleep_df['second'] = sleep_df['bedtime_time'].dt.second

# Convert the time to hours as a float
sleep_df['bedtime_hours_float'] = (sleep_df['hour'] + sleep_df['minute'] / 60 + sleep_df['second'] / 3600).round(2)

sleep_df = sleep_df.drop(columns= ['hour','minute','second','bedtime_time'])
sleep_df
```

```python
# Converting the hour into hours - integers for wakeuptime_time

# Convert the time column to datetime data type
sleep_df['wakeuptime_time'] = pd.to_datetime(sleep_df['wakeuptime_time'], format='%H:%M:%S')

# Extract the hour, minute, and second components
sleep_df['hour'] = sleep_df['wakeuptime_time'].dt.hour
sleep_df['minute'] = sleep_df['wakeuptime_time'].dt.minute
sleep_df['second'] = sleep_df['wakeuptime_time'].dt.second

# Convert the time to hours as a float
sleep_df['wakeuptime_hours_float'] = (sleep_df['hour'] + sleep_df['minute'] / 60 + sleep_df['second'] / 3600).round(2)

sleep_df = sleep_df.drop(columns= ['hour','minute','second','wakeuptime_time'])
sleep_df
```

```python
# checking the Dtype of the month columns
sleep_df.info()
```

```python
 # verify uniques in dummies for gender
sleep_df['gender'].unique()
```

```python
 # verify uniques in dummies for smoking_status
sleep_df['smoking_status'].unique()
```

```python
# change gender and smoking_status to dummies
sleep_df = pd.get_dummies(sleep_df, columns = ['gender','smoking_status'], drop_first=True)
sleep_df
```

```python
# removing id column, as we don't need it as a variable for our model
sleep_df = sleep_df.drop(columns= ['id'])
sleep_df
```

```python
# checking the Dtype of the month columns
sleep_df.info()
```

### Developing the predictive model without any transformation

```python
RAND_STATE = 34 # for reproducible shuffling
TT_RATIO = 0.3 # test/train
```

```python
y = sleep_df.sleep_efficiency
X = sleep_df.drop('sleep_efficiency', axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=TT_RATIO, random_state=RAND_STATE)
X_train = pd.DataFrame(X_train)
X_test = pd.DataFrame(X_test)
X_train.head()
```

```python
model_LR=LinearRegression()    # model
model_LR.fit(X_train, y_train)   # model train
```

```python
model_LR.coef_
```

```python
model_LR.intercept_
```

```python
y_pred = pd.DataFrame(model_LR.predict(X_test),columns = ['sleep_efficiency'] )      # model prediction
y_pred_train =  pd.DataFrame(model_LR.predict(X_train),columns = ['sleep_efficiency'])
```

```python
print(mse(y_test,y_pred))
print(mae(y_test,y_pred))
##prediction on the train set
print(mse(y_train,y_pred_train))
```

```python
R2=r2_score(y_test,y_pred)
R2
```

```python
R2_test=model_LR.score(X_test,y_test)
R2_train=model_LR.score(X_train,y_train)
Adj_R2= 1 - (1-R2)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
Adj_R2
```

```python
features_importances = pd.DataFrame(data={
    'Attribute': X_train.columns,
    'Importance':(model_LR.coef_.reshape(len(X_train.columns),)) # To check only absolute values: 'Importance':abs (model_LR.coef_.reshape(len(X_train_scaled.columns),))
})
features_importances = features_importances.sort_values(by='Importance', ascending=False)
features_importances
```

```python
plt.bar(x=features_importances['Attribute'].iloc[:15], height=features_importances['Importance'].iloc[:15], color='#087E8B')
plt.title('Feature importance rankings', size=12)
plt.xticks(rotation='vertical')
plt.show()
```

```python
X_train_const = sm.add_constant(X_train.to_numpy()) # adding a constant

model_OLS1 = sm.OLS(y_train, X_train_const).fit()
predictions_train = model_OLS1.predict(X_train_const)

X_test_const = sm.add_constant(X_test) # adding a constant
predictions_test = model_OLS1.predict(X_test_const)
print_model = model_OLS1.summary()
print(print_model)
```

```python
y_test = pd.DataFrame(y_test)
y_test = y_test.reset_index(drop = True)
```

```python
# Make an scatter plot y_pred vs y
# What kind of plot you will get if all the all the predictions are ok?
# A stright line

fig, ax = plt.subplots(1,3,figsize=(14,4))
ax[0].plot(y_pred, y_test, 'o')
ax[0].set_xlabel("y_test")
ax[0].set_ylabel("y_pred")
ax[0].set_title("Test Set -Predicted vs real")

ax[1].hist(y_test - y_pred)
ax[1].set_xlabel("Test y-y_pred")
ax[1].set_title("Test Set Residual histogram")

ax[2].plot(y_pred,y_pred.to_numpy()-y_test.to_numpy(),"o")
ax[2].set_xlabel("predited")
ax[2].set_ylabel("residuals")
ax[2].set_title("Residuals by Predicted")
ax[2].plot(y_pred,np.zeros(len(y_pred)),linestyle='dashed')
```
